```python
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Define base models
base_models = [
    ('Logistic Regression', lr),
    ('KNN', knn),
    ('SVC', svc),
    ('Decision Tree', dtc),
    ('Random Forest', rand_clf),
    ('Gradient Boosting', gbc),
    ('XGBoost', xgb)
]

# Create Voting Classifier (Hard Voting or Soft Voting)
# Use 'hard' for majority voting or 'soft' for weighted probabilities
voting_ensemble = VotingClassifier(estimators=base_models, voting='soft')

# Fit the ensemble model
voting_ensemble.fit(X_train, y_train)

# Predictions
y_pred_ensemble = voting_ensemble.predict(X_test)

# Evaluate Ensemble Model
ensemble_acc = accuracy_score(y_test, y_pred_ensemble)
print(f"Ensemble Model Accuracy: {ensemble_acc:.4f}")
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_ensemble))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_ensemble))
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression

# Define Stacking Classifier
stacking_ensemble = StackingClassifier(
    estimators=base_models,
    final_estimator=LogisticRegression()  # Meta-model
)

# Fit and evaluate
stacking_ensemble.fit(X_train, y_train)
y_pred_stacking = stacking_ensemble.predict(X_test)

# Evaluation
stacking_acc = accuracy_score(y_test, y_pred_stacking)
print(f"Stacking Model Accuracy: {stacking_acc:.4f}")
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_stacking))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_stacking))
```

```python
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt

models = [
    {
        'label': 'LR',
        'model': lr,
    },
    {
        'label': 'DT',
        'model': dtc,
    },
    {
        'label': 'SVM',
        'model': svc,
    },
    {
        'label': 'KNN',
        'model': knn,
    },
    {
        'label': 'XGBoost',
        'model': xgb,
    },
    {
        'label': 'RF',
        'model': rand_clf,
    },
    {
        'label': 'GBDT',
        'model': gbc,
    },
    {
        'label': 'Ensemble',
        'model': voting_ensemble,
    }
]

means_roc = []
means_accuracy = [
    100 * round(lr_acc, 4),
    100 * round(dtc_acc, 4),
    100 * round(svc_acc, 4),
    100 * round(knn_acc, 4),
    100 * round(xgb_acc, 4),
    100 * round(rand_clf_acc, 4),
    100 * round(gbc_acc, 4)
]

# Compute accuracy and ROC for each model
for m in models:
    model = m['model']
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    fpr1, tpr1, thresholds = metrics.roc_curve(y_test, model.predict_proba(X_test)[:, 1])
    auc = metrics.roc_auc_score(y_test, model.predict(X_test))
    auc = 100 * round(auc, 4)
    means_roc.append(auc)

# Append the ensemble accuracy to means_accuracy
ensemble_acc = 100 * round(voting_ensemble.score(X_test, y_test), 4)
means_accuracy.append(ensemble_acc)

print(means_accuracy)
print(means_roc)

# Data to plot
n_groups = len(models)  # Update number of groups to include Ensemble
means_accuracy = tuple(means_accuracy)
means_roc = tuple(means_roc)

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))  # Adjusted size for better clarity
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, means_accuracy, bar_width,
                 alpha=opacity,
                 color='mediumpurple',
                 label='Accuracy (%)')

rects2 = plt.bar(index + bar_width, means_roc, bar_width,
                 alpha=opacity,
                 color='rebeccapurple',
                 label='ROC (%)')

# Adjust limits and labels
plt.xlim([-1, n_groups])
plt.ylim([40, 100])  # Adjusted limit to accommodate all scores
plt.title('Performance Evaluation - Liver Disease Prediction', fontsize=12)
plt.xticks(index, ('LR', 'DT', 'SVM', 'KNN', 'XGBoost', 'RF', 'GBDT', 'Ensemble'),
           rotation=40, ha='center', fontsize=12)
plt.legend(loc="upper right", fontsize=10)

# Save and show the plot
plt.savefig("PE_liver_with_ensemble.jpeg", format='jpeg', dpi=400, bbox_inches='tight')
plt.show()
```
